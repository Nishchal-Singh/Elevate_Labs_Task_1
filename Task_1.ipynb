# Importing Library

```python
import pandas as pd
```

# Importing Dataset Medical Appointment No Shows

```python
df = pd.read_csv("KaggleV2-May-2016.csv")
```

# 1 Handling missing value

```python
df = df.dropna()
```

# 2 Removing duplicates

```python
df = df.drop_duplicates()
```

# 3 Standardize text values

```python
df['Gender'] = df['Gender'].str.strip().str.lower()
df['Gender'] = df['Gender'].replace({'f': 'Female', 'm': 'Male'})
```

```python
# For 'No-show' column: convert 'No'/'Yes' to lowercas
```

```python
if 'No-show' in df.columns:
    df['No-show'] = df['No-show'].str.strip().str.lower()
```

# 4 Convert date formats to a consistent type

```python
for col in df.columns:
    if 'date' in col.lower():
        df[col] = pd.to_datetime(df[col], errors='coerce')
```

# 5 Rename column headers to be clean and uniform

```python
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_").str.replace("-", "_")
```

# 6 Check and fix data types

```python
if 'age' in df.columns:
    df['age'] = df['age'].astype(int)
```

```python
df.to_csv("KaggleV2-May-2016-cleaned.csv", index=False)
```
